```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.random.randint(100,500,100)
y = np.random.randint(100,1000,100)
y1 = np.random.randint(100,1000,100)
```

```python
plt.scatter(x,y, marker='s',color='red')
plt.scatter(x,y1, marker='^',color='blue')
plt.title("Simple-Scatter")
plt.xlabel('X---->')
plt.ylabel('Y---->')
```

```python
x_ = np.arange(10)
y_ = np.random.randint(30,100,10)
y1_ = np.random.randint(30,100,10)
```

```python
xlist = ["sub1","sub2","sub3","sub4","sub5","sub6","sub7","sub8","sub9","sub10"]
plt.xticks(x_,xlist)
plt.bar(x_,y_,color='red')
plt.bar(x_,y1_,color='blue',bottom=y_)
plt.title("Bar-Graph")
plt.xlabel("Subjects")
plt.ylabel("Marks obtained")
plt.show()
```

```python
x__ = np.arange(1,7)
y__ = np.random.randint(50,100,6)
```

```python
xlist = ["sem1","sem2","sem3","sem4","sem5","sem6"]
plt.xticks(x__,xlist)
plt.title("Frequency-Polygon-Graph")
plt.xlabel("sem")
plt.ylabel("performance")
plt.plot(x__,y__,"b^")
plt.plot(x__,y__,"red")
```
